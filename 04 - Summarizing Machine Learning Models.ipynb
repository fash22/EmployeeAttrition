# Project Phase 4: Summarizing Machine Learning Models
With preprocessed data, we can now conduct machine learning modelling. This notebook summarizes the results of the individual machine learning modelling. These modelling can be found at the `./models` directory.

## Importing the Results of the Models
Let's start by importing pandas, and use it to load the results from our models. These results were saved at `./dataset/model-results` directory.

```python
import pandas as pd
import seaborn as sns

# Task: Set Color Pallete
PALLETE = sns.color_palette("Accent")
sns.set_palette(PALLETE)

model_results = {
  "Logistic Regression"       : pd.read_csv("dataset/model-results/LogisticRegression.csv"),
  "Support Vector Classifier" : pd.read_csv("dataset/model-results/SupportVectorClassifier.csv"),
  "Random Forest Classifier"  : pd.read_csv("dataset/model-results/RandomForestClassifier.csv"),
  "Neural Network1"           : pd.read_csv("dataset/model-results/NeuralNetwork1.csv")
}
```

## Tabulating the Results
Once imported, let's tabulate the results of the model so that we can compare them side by side.

```python
results = pd.DataFrame({
  "Model": model_results.keys(),
  "Accuracy Score": [model_results[result].iloc[0,2] for result in model_results],
  "Precision Score": [model_results[result].iloc[1,2] for result in model_results],
  "Recall Score": [model_results[result].iloc[2,2] for result in model_results],
  "F1 Score": [model_results[result].iloc[3,2] for result in model_results],
})
```

## Comparing Model Accuracies

```python
sns.barplot(data=results, y="Model", x="Accuracy Score", orient="h");
```

```python
sns.barplot(data=results, y="Model", x="Precision Score", orient="h");
```

```python
sns.barplot(data=results, y="Model", x="Recall Score", orient="h");
```

```python
sns.barplot(data=results, y="Model", x="F1 Score", orient="h");
```
